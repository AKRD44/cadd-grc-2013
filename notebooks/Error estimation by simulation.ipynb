# Error estimation by simulation

Suppose we have a dataset $x_i$, $i = 1,\ldots,N$, along with corresponding uncertainties $\delta x_i$.  We assume the error for each sample is Gaussian.

```python
import numpy
from numpy.random import uniform, exponential
nsamples = 6;
x_i = uniform(-3, +3, size=[nsamples]); # measurements
dx_i = 0.5 + exponential(0.05, size=[nsamples]); # measurement errors
```

We can visualize the uncertainty in each measurement.

```python
from pylab import *
clf();
hold(True);
xpoints = linspace(-5, +5, 100);
from numpy import sqrt, exp, pi
for i in range(nsamples):    
    plot(xpoints, 1.0/(sqrt(2*pi)*dx_i[i])*exp(-(xpoints-x_i[i])**2/(2*dx_i[i]**2)), 'k-')
xlabel('x');
```

Suppose we have a complex function $f(\{x_i\})$ of the measurements, such as
$$f(\{x_i\}) = \log \sum_{i=1}^N e^{x_i}$$

```python
def f(x_i):
    from numpy import log, exp, sum
    return log(sum(exp(x_i)))
```

To estimate the error in the function $f$ of the data, we can *simulate* artificial measurements using the measured values $x_i$ and their standard errors $\delta x_i$ using a Gaussian model for the uncertainty in each measurement.

```python
def simulate_uncertainty(f, x_i, dx_i, nreplicates=10000):
    "Compute the uncertainty of a function f(...) of a dataset x_i and its uncertainties dx_i by simulation."
    from numpy.random import standard_normal
    f_n = numpy.zeros([nreplicates], numpy.float64) 
    N = x_i.size
    for replicate in range(nreplicates):
        # Generate new sample.
        x_sample = x_i + dx_i * standard_normal(size=[N])
        # Compute function on bootstrapped sample.
        f_n[replicate] = f(x_sample)
    return f_n
```

```python
# Compute uncertainty distribution of f by simulation.
f_n = simulate_uncertainty(f, x_i, dx_i)

# Plot uncertainty distribution of f.
nbins = 40;
hist(f_n,nbins);
xlabel('f');
ylabel('p(f)');
gca().axes.get_yaxis().set_ticks([]); # turn off y ticks

# Label plot with mean and standard error.
F = f_n.mean();
dF = f_n.std();
title('F = %.1f $\pm$ %.1f' % (F, dF));
```
