# Propagation of error

Suppose we have a function $f(x)$ of some measurement $x$ that is treated as a random variable.

For concreteness, let's consider a binding free energy $\Delta G$ computed from a dissociation constant $K_d$ at $T = 300$ K:

$$\Delta G = k_B T \ln (K_d C_0)$$
                        
Here, $k_B$ is the Boltzmann constant and $C_0$ is the standard state concentration (1 M).

```python
import numpy
from numpy.random import standard_normal

k_B = 1.381e-23 * 6.022e+23 / 4184.0 # Boltzmann constant (J/K)
T = 300 # temperature (K)
C_0 = 1.0 # standard state concentration (M)

def f(K_d):
    return k_B * T * numpy.log(K_d * C_0)
```

Now suppose we have measured the $K_d$ to be 1.0$\pm$0.1 uM, with the error being well-represented by a Gaussian.

```python
K_d = 1.0e-6 # measured dissociation constant (M)
dK_d = 0.1e-6 # uncertainty in measured dissociation constant (M)

nsamples = 50000
x = K_d + dK_d * standard_normal([nsamples])
y = f(x)
```

```python
from matplotlib.pylab import *
nbins = 50
hist(x, nbins);
xlabel('$K_d$ (M)');
ylabel('p($K_d$)');

# fancy plot formatting
rc('font', family='sans-serif', size=15); # increase font size
gca().axes.get_yaxis().set_ticks([]); # turn off y ticks
xticks([0.6e-6, 0.8e-6, 1.0e-6, 1.2e-6, 1.4e-6], ['$0.6 \cdot 10^{-6}$', '$0.8 \cdot 10^{-6}$', '$1.0 \cdot 10^{-6}$', '$1.2 \cdot 10^{-6}$', '$1.4 \cdot 10^{-6}$']);
```

The binding free energy is also approximately Gaussian, centered on $\Delta G = k_B T \log (K_d C_0)$.  

```python
hist(y, nbins);
xlabel('$\Delta G$ (kcal/mol)');
ylabel('$p(\Delta G)$');
gca().axes.get_yaxis().set_ticks([]); # turn off y ticks
```

There are clearly higher-order moments that are non-Gaussian, but in many cases, the Gaussian approximation works quite well.

The uncertainty (standard deviation) of the new distribution is modified by the functional transformation $y = f(x)$.  We can compute the effect applicaiton of $f(x)$ on the uncertainty---what is generally called "propagating the error"---by expanding $f(x)$ in a first-order Taylor series:
    $$f(x_0 + \Delta x) = f(x_0) + \left.\frac{\partial f}{\partial x}\right|_{x_0} \Delta x + \mathcal{O}(\Delta x^2)$$
Rearranging, we see that the error in $f$, $\Delta f$, to first order, is
    $$\Delta f \equiv f(x_0 + \Delta x) - f(x_0) = \left.\frac{\partial f}{\partial x}\right|_{x_0} \Delta x + \mathcal{O}(\Delta x^2)$$
The squared-uncertainty is the variance of $\Delta f$, which is
    $$\delta^2 f = E[\Delta f^2] = E\left[ \left( \left.\frac{\partial f}{\partial x}\right|_{x_0} \Delta x\right)^2 \right] = \left( \left.\frac{\partial f}{\partial x}\right|_{x_0} \right)^2 \delta^2 x$$
The uncertainty is just the square root of this: 
    $$\delta f = \left|\frac{\partial f}{\partial x}\right|^2_{x_0} \delta x$$


For the transformation to binding free energies
$$\Delta G = k_B T \ln (K_d C_0)$$
we can then compute the uncertainty in $\Delta G$ given the measured uncertainty in $K_d$,
$$\delta \Delta G = \left| \frac{k_B T}{K_d C_0} \right| C_0 \delta K_d = k_B T \frac{\delta K_d}{K_d}$$


```python
DeltaG = k_B*T*numpy.log(K_d*C_0) # Compute DeltaG measurement from K_d measurement
dDeltaG = k_B*T*(dK_d/K_d) # propagate error in K_d into error in DeltaG using first-order Taylor series approximation

[Nbins, xbins, patches] = hist(y, nbins);
hold(True);
dx = xbins[2] - xbins[1] # grid spacing
plot(xbins, (nsamples*dx)/(sqrt(2*pi)*dDeltaG) * exp(-(xbins-DeltaG)**2 / (2*dDeltaG**2)), 'r-', linewidth=2);

# label and adjust plot
xlabel('$\Delta G$ (kcal/mol)');
ylabel('$p(\Delta G)$');
gca().axes.get_yaxis().set_ticks([]); # turn off y ticks
legend(['error propagation', 'actual distribution'], 'upper left', fontsize=12);
```

The Taylor series propagation of error and Gaussian error approximation work quite well in this case, even though the function $f(x)$ was highly nonlinear!

For functions $f(x_1, x_2, \ldots, x_N)$ of multiple variables $x_1, x_2, \ldots, x_N$, applying the same Taylor expansion procedure (now multidimensional) leads to the generalized propagation of error expression for the squared-error:
    $$\delta^2 f = \sum_{i=1}^{N} \sum_{j=1}^N \left. \frac{\partial f}{\partial x_i} \frac{\partial f}{\partial x_j} \right|_{x_1,\ldots,x_N} \delta x_i \delta x_j$$
where $\delta x_i \delta x_j$ is the covariance or correlated error between measurements $x_i$ and $x_j$.  If the measurements $x_i$ and $x_j$ are independent (fully independent measurements), then $\delta x_i \delta x_j = 0$.  If $i = j$, then $\delta x_i \delta x_j = \delta^2 x_i$ is the squared-error of measurement $x_i$.

## Representing uncertainties with a box-and-whisker plot

Here's an example of a notched box-and-whisker plot.

```python
boxplot(y, vert=True, notch=True);
ylabel('$\Delta G$ (kcal/mol)');
xlabel('measurement');
```

## Breakdown of first-order Taylor expansion based error propagation

Because the error propagation scheme above omits all terms higher order than the linear term, there are cases where it will break down when these terms become non-negligible.  This often occurs when there are nearby boundaries or limits on the values the random variable or measurement can assume.

For example, let's assume we measure a quantity $x$ that must be somewhere in the interval $[0,1]$.  Let the uncertainty have a Beta distribution, which appropriately reflects the fact that $x \in [0,1]$:
$$x \sim \mathrm{Beta}(\alpha, \beta)$$

```python
from numpy.random import beta
a = 8; b = 2; nsamples = 50000;
x_n = beta(a, b, size=[nsamples]);
[Nbins, xbins, patches] = hist(x_n, nbins);
xlabel('x');
ylabel('p(x)');
gca().axes.get_yaxis().set_ticks([]); # turn off y ticks

hold(True);
x = x_n.mean(); # measured value
dx = x_n.std(); # uncertainty from standard deviation
bin_width = xbins[1] - xbins[0];
plot(xbins, (nsamples*bin_width)/(sqrt(2*pi)*dx) * exp(-(xbins-x)**2 / (2*dx**2)), 'r-', linewidth=2);

legend(['actual', 'Gaussian'], 'upper left');
```

If we now compute a transformation of this function such as $y = \log(x)$, the Gaussian error assumption and Taylor error propagtion does not work well.

```python
y_n = numpy.log(x_n);
[Nbins, ybins, patches] = hist(y_n, nbins);
xlabel('y');
ylabel('p(y)');
title('$y = \log x$');
gca().axes.get_yaxis().set_ticks([]); # turn off y ticks

# Propagate error:
y = numpy.log(x); # y = log(x)
dy = (dx/x); # first-order Taylor propagation of error

hold(True);
bin_width = ybins[1] - ybins[0];
plot(ybins, (nsamples*bin_width)/(sqrt(2*pi)*dy) * exp(-(ybins-y)**2 / (2*dy**2)), 'r-', linewidth=2);

legend(['actual', 'error propagation'], 'upper left');
```
