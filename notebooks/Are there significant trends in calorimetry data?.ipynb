# Are there significant trends in the enthalpy as a function of time?

Consider the HIV-1 protease dataset compiled in a recent paper:

```python
import numpy
DH = numpy.array([1.8, 1.2, 3.1, -4.3, -6.9, -3.8, -4.2, -0.7, -12.7])
mTDS = numpy.array([-14.2, -14.2, -15.9, -9.4, -6.3, -11.3, -10.1, -13.9, -2.3])
DG = numpy.array([-12.4, -13.0, -12.8, -13.7, -13.2, -15.1, -14.3, -14.6, -15.0])
indices = numpy.arange(DH.size);
compounds = ['indinavir', 'saquinavir', 'nelfinavir', 'ritonavir', 'amprenavir', 'lopinavir', 'atazanavir', 'tipranavir', 'darunavir']
```

```python
clf();
hold(True);
plot(DH, 'go');
plot(DG, 'bo');
plot(mTDS, 'ro');
xlabel('time'); ylabel('kcal/mol');
```

```python
clf();
bar(indices, DH);
ylabel('$\Delta H$ (kcal/mol)');
```

```python
clf()
bar(indices, DG);
ylabel('$\Delta G$ (kcal/mol)');
```

```python
clf();
bar(indices, mTDS);
ylabel('$-T \Delta S$ (kcal/mol)');
```

Let's fit a linear trend to the enthalpy data.

```python
clf();
x = indices
y = DH
plot(x, y, 'go');
ylabel('$\Delta H$ (kcal/mol)');
```

```python
# regress using y' = alpha + beta * x
from numpy import mean, var, cov
beta = cov(x,y)[0,1] / var(x)
alpha = mean(y) - beta * mean(x)
```

```python
# Plot the data and our linear regression to the data.
clf();
plot(x, y, 'go', indices, alpha + beta * indices, 'g-');
xlabel('x'); ylabel('y');
title("Slope is %.3f" % beta);
```

```python
# Define a linear regression function.
def linear_regression(x, y):
    beta = cov(x,y)[0,1] / var(x)
    alpha = mean(y) - beta * mean(x)
    yprime = alpha + beta * x
    R2 = 1 - sum((yprime-y)**2) / sum((y-mean(y))**2)
    return [alpha, beta, yprime, R2]    

# Estimate R2.
[alpha, beta, yprime, R2] = linear_regression(x, y)

# Bootstrap uncertainty estimation
from numpy import arange
from numpy.random import choice
ntrials = 1000
nsamples = x.size
alpha_n = numpy.zeros([ntrials], numpy.float64)
beta_n = numpy.zeros([ntrials], numpy.float64)
R2_n = numpy.zeros([ntrials], numpy.float64)
for trial in range(ntrials):
    # Generate sample with replacement.
    bootstrap_sample_indices = choice(arange(nsamples), [nsamples], replace=True)
    x_sample = x[bootstrap_sample_indices]
    y_sample = y[bootstrap_sample_indices]
    # Perform regression on sample.
    [alpha_sample, beta_sample, yprime_sample, R2_sample] = linear_regression(x_sample, y_sample)
    # Compute R^2 for sample.
    R2_n[trial] = R2_sample
    # Store alpha and beta.
    alpha_n[trial] = alpha_sample
    beta_n[trial] = beta_sample
# Compute uncertainty in R^2.
dR2 = R2_n.std()
dalpha = alpha_n.std()
dbeta = beta_n.std()
print "R2 = %.5f +- %.5f" % (R2, dR2)
print "slope = %.5f +- %.5f" % (beta, dbeta)
```

```python
clf();
plot(x, y, 'go', indices, alpha + beta * indices, 'g-');
xlabel('x'); ylabel('y');
title("Slope is %.3f $\pm$ %.3f" % (beta, dbeta));
hold(True);
for trial in range(ntrials):
    plot(indices, alpha_n[trial] + beta_n[trial] * indices, 'g-')
```
